```python
import numpy as np
import pandas as pd
import matplotlib as plt
df = pd.read_csv(r"student_depression_dataset.csv",index_col= 0,header = 0 )
```

```python
print(df.shape)
```

```python
def sleep_fix(x: str):
    encounters = []
    temp = ""
    for y in range(len(x)):
        if x[y].isdigit():
            if len(temp) > 0:
                temp = temp + x[y]
                encounters.append(str(temp))
            else:
                temp = x[y]
            
        else :
            if len(temp) > 0 :
                encounters.append(str(temp))
            temp = ""
    if len(encounters) > 0 :
        return max(encounters)
    else:
        return ""
df['Sleep Duration'] = df['Sleep Duration'].apply(sleep_fix)
```

```python
def remove_unwanted_cities(x:str):
    unwanted = ['Saanvi', 'M.Tech', 'Bhavna', 'City', '3',
       "'Less than 5 Kalyan'", 'Mira', 'Harsha', 'Vaanya', 'Gaurav',
       'Harsh', 'Reyansh', 'Kibara', 'Rashi', 'ME', 'M.Com', 'Nalyan',
       'Mihir', 'Nalini', 'Nandini']
    if x in unwanted:
        return pd.NA
    elif x == 'Khaziabad' :
        return 'Ghaziabad'
    else:
        return x
df['City'] = df['City'].apply(remove_unwanted_cities)
```
